```python
def f(x):
    #print(type(x))   // x is Series which are divided by DataFrame
    #print(type(x.max()))
    #print(x)
    Max = x.max()
    secondLarget = -1000000
    for ele in x:
        if ele>secondLarget and ele!=Max:
            secondLarget = ele
    return Series([x.max(), x.min(), secondLarget] ,index=['max', 'min', 'secondLarget'])
```

```python
temp = data1.iloc[:,[13,14,15,17,18]]
temp.apply(f)
temp.corr()
```
