# 1. List

```python
def insertNumToList(num, ls):
    oldList = ls.copy()
    print("The old and new lists are: ")
    for element in ls:
        if num <= element:
            position = ls.index(element)
            ls.insert(position, num)
            return oldList, ls
    ls.append(num)       # num is bigger than all the elements in the list
    return oldList, ls

ls = [1 , 2 , 4 , 9 , 17 , 25 , 53 ]
num = 13
insertNumToList(num, ls)
        
```

# 2. Bounce Height

```python
def calHeight(original, bounce):
    if original < 0 :
        raise Exception('The height must be positive')
    if bounce < 0 or bounce != int(bounce) :
        raise Exception('The bouncing times must be positive integer')
    
    height = original
    i = bounce
    while(i>1):
        height += original
        original /= 2
        i -= 1
    original /= 2
    height += original
    print("After %d bounce, the ball bounces back to %.4f meters height\n" %(bounce, original) )
    return height
            

calHeight(138, 5)
        
    
```
